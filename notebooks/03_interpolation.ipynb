# Latent interpolation

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

from gen_cats.evaluation.report_analysis import (
    ensure_plots_dir,
    list_interpolation_images,
    write_stats_json,
)

PLOTS = ensure_plots_dir("results")
images = list_interpolation_images()
images
```

```python
def panel_label(path) -> str:
    return path.stem.replace("interpolation_", "").replace("_", " ")


imgs = [mpimg.imread(path) for path in images]
fig_width = 10.5
title_space = 0.24
row_gap = 0.08
strip_heights = [fig_width / (img.shape[1] / img.shape[0]) for img in imgs]
fig_height = sum(strip_heights) + len(imgs) * title_space + (len(imgs) - 1) * row_gap

fig, axes = plt.subplots(len(imgs), 1, figsize=(fig_width, fig_height), squeeze=False)

for ax, path, img in zip(axes[:, 0], images, imgs, strict=True):
    ax.imshow(img, aspect="auto")
    ax.set_title(panel_label(path), loc="left", fontsize=11, pad=5)
    ax.axis("off")

fig.subplots_adjust(hspace=0.40, left=0.002, right=0.998, top=0.99, bottom=0.002)
out = PLOTS / "interpolation_panel.png"
fig.savefig(out, dpi=200, bbox_inches="tight", pad_inches=0.01)
plt.close(fig)
out
```

```python
write_stats_json(
    "interpolation_index.json",
    {
        "n_images": len(images),
        "files": [p.name for p in images],
        "models": sorted(
            {"_".join(p.stem.replace("interpolation_", "").split("_")[:-1]) for p in images}
        ),
        "seeds": sorted({int(p.stem.split("_seed")[-1]) for p in images}),
    },
)
```
