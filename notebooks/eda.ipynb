```python
import matplotlib.pyplot as plt
import numpy as np
import PIL.Image as Image
import seaborn as sns

sns.set_style("whitegrid")
sns.set_palette("Set2")
```

# CATS

```python
train = np.load("../data/processed/train.npy")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(7, 7))

for i, ax in enumerate(axes.flat):
    img = train[i]

    ax.imshow(img)
    ax.set_title("")
    ax.axis("off")

plt.tight_layout()
plt.savefig("./plots/processed_samples.png", dpi=500)
plt.show()
```

```python
after_processing = train[3]
before_processing = Image.open("../data/raw/cats/00000001_011.jpg")

before_processing = before_processing.resize((128, 128))
```

```python
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(before_processing)
plt.title("Before processing")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(after_processing)
plt.title("After processing")
plt.axis("off")

plt.tight_layout()
plt.savefig("./plots/preprocessing_comparison.png", dpi=200)
plt.show()
```
